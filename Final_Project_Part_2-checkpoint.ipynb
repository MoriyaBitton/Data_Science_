**Question 1**

```python
def decimal_to_binary(dec): 
    decimal = int(dec) 
  
    print(decimal, " in Binary : ", bin(decimal)) 
    
def decimal_to_octal(dec): 
    decimal = int(dec) 
  
    print(decimal, "in Octal : ", oct(decimal)) 

def decimal_to_hexadecimal(dec): 
    decimal = int(dec) 
  
    print(decimal, " in Hexadecimal : ", hex(decimal)) 

dec = input("Enter a number : ")
decimal_to_binary(dec) 
decimal_to_octal(dec) 
decimal_to_hexadecimal(dec)
```

**Question 2**

```python
import pandas as pd
cast = pd.read_csv('cast.csv')
cast.head(10)
```

**Question 2.1**

How many movies have the title "Hamlet"?

```python
hamletMovies = cast[cast["title"] == "Hamlet"]
hamletMovies2 = hamletMovies.drop_duplicates('year', keep='first').reindex()
print("The number of films named \"Hamlet\" is:",hamletMovies2.shape[0])
```

**Question 2.2** 

List all of the "Treasure Island" movies from earliest to most recent.

```python
numTI = cast[cast["title"] == "Treasure Island"].drop_duplicates('year', keep='first').reindex()
numTI = numTI.drop(["name", "type", "character", "n"], axis = 1)
numTI.sort_values("year").reset_index(0, drop=True)
```

**Question 2.3**

How many roles were credited in the silent 1921 version of Hamlet?

```python
hamlet1921 = hamletMovies.loc[cast["year"] == 1921].shape[0]
print("The number of roles were credited in the silent 1921 version of \"Hamlet\" is:", hamlet1921)
```

**Question 2.4**

Use groupby() to plot the number of "Hamlet" films made each decade

```python
hamletDecade = hamletMovies.groupby("year").sum().reset_index(0, drop=False)
hamletDecade["decade"] = (10 * (hamletDecade["year"] // 10)).astype(str) + 's'
hamletDecade['Table of Hamlet movie\'s per decade'] = 1
hamletDecade = hamletDecade.drop(['n', "year"], axis=1).groupby("decade").sum()
hamletDecade
```

**Question 2.5**

How many leading (n=1) roles were available to actors, and how many to actresses, in each year of the 1950s?

```python
cast_50s = cast.loc[cast["year"] < 1960].loc[cast["year"] > 1949]
cast_50s_n1 = cast_50s.loc[cast["n"] == 1]
actor_50s = cast_50s_n1.loc[cast["type"] == "actor"]
actress_50s = cast_50s_n1.loc[cast["type"] == "actress"]

print("The number of leading roles in the 1950's")
print("for actors:", actor_50s.shape[0])
print("for actress:", actress_50s.shape[0])
```

**Question 2.6**

List the 10 actors/actresses that have the most leading roles (n=1) since the 1990's.

```python
actor_90s = cast.loc[cast["year"] > 1989].loc[cast["n"] == 1]
actor_90s = actor_90s.drop(['n', "year"], axis=1)
actor_90s["roles"] = 1
actor_90s = actor_90s.groupby(["name", "type"]).sum().reset_index(0, drop=False).reset_index(0, drop=False)
actor_90s.sort_values("roles", ascending=False).reset_index(0, drop=True).head(10)
```

**Question 2.7**

List, in order by year, each of the films in which Frank Oz has played more than 1 role

```python
frankOz = cast.loc[cast["name"] == "Frank Oz"].sort_values("year")
frankOz["roles"] = 1
frankOz = frankOz.drop(['n', "type"], axis=1).groupby(["title", "year"]).agg({'roles':'sum'})
frankOz = frankOz.sort_values("year").loc[frankOz["roles"] > 1].reset_index(0, drop=False).reset_index(0, drop=False)
frankOz
```
